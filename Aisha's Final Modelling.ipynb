```
def cap_floor(series, lower_q=0.01, upper_q=0.99):
    lower = series.quantile(lower_q)
    upper = series.quantile(upper_q)
    return np.clip(series, lower, upper)

for col in df_dropped.drop(columns=['diagnosis']):
    df_dropped[col] = cap_floor(df_dropped[col])
```

```
import statsmodels.api as sm

def forward_selection(X, y, sl_enter=0.05):
    X_const = sm.add_constant(X)
    features = ['const']
    remaining = list(X_const.columns)
    remaining.remove('const')

    while True:
        best_pval = float('inf')
        best_feature = None
        for feature in remaining:
            model = sm.Logit(y, X_const[features + [feature]]).fit(disp=0)
            pval = model.pvalues[feature]
            if pval < best_pval:
                best_pval = pval
                best_feature = feature
        if best_pval < sl_enter:
            features.append(best_feature)
            remaining.remove(best_feature)
        else:
            break

    final_model = sm.Logit(y, X_const[features]).fit()
    return final_model, features
```

```
def backward_elimination(X, y, sl_remove=0.05):
    X_const = sm.add_constant(X)
    features = list(X_const.columns)
    while True:
        model = sm.Logit(y, X_const[features]).fit(disp=0)
        pvalues = model.pvalues
        max_pval = pvalues.drop('const', errors='ignore').max()
        if max_pval > sl_remove:
            worst_feature = pvalues.drop('const', errors='ignore').idxmax()
            features.remove(worst_feature)
        else:
            break
    final_model = sm.Logit(y, X_const[features]).fit()
    return final_model, features
```

```
def stepwise_selection(X, y, sl_enter=0.05, sl_remove=0.05):
    X_const = sm.add_constant(X)
    included = ['const']
    remaining = list(X_const.columns)
    remaining.remove('const')

    while True:
        changed = False

        # Forward step
        best_feature = None
        best_pval = float('inf')
        for feature in remaining:
            model = sm.Logit(y, X_const[included + [feature]]).fit(disp=0)
            pval = model.pvalues[feature]
            if pval < best_pval:
                best_pval = pval
                best_feature = feature
        if best_pval < sl_enter:
            included.append(best_feature)
            remaining.remove(best_feature)
            changed = True

        # Backward step
        model = sm.Logit(y, X_const[included]).fit(disp=0)
        pvalues = model.pvalues.drop('const', errors='ignore')
        worst_pval = pvalues.max()
        if worst_pval > sl_remove:
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            remaining.append(worst_feature)
            changed = True

        if not changed:
            break

    final_model = sm.Logit(y, X_const[included]).fit()
    return final_model, included
```

```
# Forward
fwd_model, fwd_features = forward_selection(X_train, y_train)
print("Forward Selected Features:", fwd_features)
print(fwd_model.summary())

# Backward
bwd_model, bwd_features = backward_elimination(X_train, y_train)
print("Backward Selected Features:", bwd_features)
print(bwd_model.summary())

# Stepwise
step_model, step_features = stepwise_selection(X_train, y_train)
print("Stepwise Selected Features:", step_features)
print(step_model.summary())
```

```
y_pred_fwd_test_prob = fwd_model.predict(X_test_const[fwd_features])
y_pred_fwd_test = (y_pred_fwd_test_prob >= 0.5).astype(int)

print("\nðŸ”· Forward Selection - Test Classification Report:")
print(classification_report(y_test, y_pred_fwd_test))

# Predictions on Train Set
y_pred_fwd_train_prob = fwd_model.predict(X_train_const[fwd_features])
y_pred_fwd_train = (y_pred_fwd_train_prob >= 0.5).astype(int)

print("\nðŸ”· Forward Selection - Train Classification Report:")
print(classification_report(y_train, y_pred_fwd_train))
```

```
# Predictions on Test Set
y_pred_bwd_test_prob = bwd_model.predict(X_test_const[bwd_features])
y_pred_bwd_test = (y_pred_bwd_test_prob >= 0.5).astype(int)

print("\nðŸ”¶ Backward Elimination - Test Classification Report:")
print(classification_report(y_test, y_pred_bwd_test))

# Predictions on Train Set
y_pred_bwd_train_prob = bwd_model.predict(X_train_const[bwd_features])
y_pred_bwd_train = (y_pred_bwd_train_prob >= 0.5).astype(int)

print("\nðŸ”¶ Backward Elimination - Train Classification Report:")
print(classification_report(y_train, y_pred_bwd_train))
```

```
# Predictions on Test Set
y_pred_step_test_prob = step_model.predict(X_test_const[step_features])
y_pred_step_test = (y_pred_step_test_prob >= 0.5).astype(int)

print("\nðŸŸ© Stepwise Selection - Test Classification Report:")
print(classification_report(y_test, y_pred_step_test))

# Predictions on Train Set
y_pred_step_train_prob = step_model.predict(X_train_const[step_features])
y_pred_step_train = (y_pred_step_train_prob >= 0.5).astype(int)

print("\nðŸŸ© Stepwise Selection - Train Classification Report:")
print(classification_report(y_train, y_pred_step_train))
```
